```python
import numpy as np
import matplotlib.pyplot as plt
# なければpip install japanize_matplotlibして(グラフを日本語フォントに勝手に対応してくれるライブラリ)
import japanize_matplotlib

def q1(N):
    # 遷移行列の生成
    matrix_s = []
    for i in range(0,N+1):
        elem_matrix_s = []
        for j in range(0,N+1):
            if j == i:
                elem_matrix_s.append(i/N)
            elif j == i+1:
                elem_matrix_s.append((N-i)/N)
            else:
                elem_matrix_s.append(0)
        matrix_s.append(elem_matrix_s)

    matrix_trans = np.array(matrix_s)
    print("遷移行列A = \n{}".format(matrix_trans))

# N = おまけの種類, n = 買った回数(試行回数)
def q2(N, n):
    # 遷移行列の生成
    matrix_s = []
    for i in range(0,N+1):
        elem_matrix_s = []
        for j in range(0,N+1):
            if j == i:
                elem_matrix_s.append(i/N)
            elif j == i+1:
                elem_matrix_s.append((N-i)/N)
            else:
                elem_matrix_s.append(0)
        matrix_s.append(elem_matrix_s)

    matrix_trans = np.array(matrix_s)
    
    # 初期ベクトルP(0) [1,0,...,0]
    vec_p = np.zeros(N+1)
    vec_p[0] = 1
    
    # グラフ描画
    x = []
    y = []
    vec_p_before = 0
    for i in range(n):
        vec_p = np.dot(vec_p, matrix_trans)
        x.append(i)
        y.append(vec_p[-1] - vec_p_before)
        vec_p_before = vec_p[-1]
    plt.xlabel("試行回数n[回]")
    plt.ylabel("初めてN種類揃う確率pN(n)")
    plt.plot(x,y)
    plt.show()
        
def q3(N, n):
    # 遷移行列の生成
    matrix_s = []
    for i in range(0,N+1):
        elem_matrix_s = []
        for j in range(0,N+1):
            if j == i:
                elem_matrix_s.append(i/N)
            elif j == i+1:
                elem_matrix_s.append((N-i)/N)
            else:
                elem_matrix_s.append(0)
        matrix_s.append(elem_matrix_s)

    matrix_trans = np.array(matrix_s)
    
    # 初期ベクトルP(0) [1,0,...,0]
    vec_p = np.zeros(N+1)
    vec_p[0] = 1
    
    # グラフ描画
    x = []
    y = []
    for i in range(n):
        vec_p = np.dot(vec_p, matrix_trans)
        x.append(i)
        y.append(vec_p[-1])
    plt.xlabel("試行回数n[回]")
    plt.ylabel("N種類揃う累計確率pN(n)")
    plt.plot(x,y)
    plt.show()
        
q1(10)
q2(10,100)
q3(10,100)
```
